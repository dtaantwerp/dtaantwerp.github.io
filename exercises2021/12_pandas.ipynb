#### Exercise 1: Write a class for a house
- The class should contain at least three attributes
    - Type of floor
    - Colour of the walls
    - Number of rooms
- The class should contain at least three methods
    - Paint walls
    - Renovate floor
    - Regulate temperature
    
Once the class is written,
- Create a class object 
- Print the attributes
- Paint the walls to another color
- Renovate the floor
- Regulate the temperature

```python
class House():
    floor = 'parquet'
    colour_paint = 'white'
    n_rooms = 10
    
    def paint_walls(self, new_colour):
        self.colour_paint = new_colour
        
    def renovate_floor(self, new_floor):
        self.floor = new_floor
        
    def regulate_temp(self, temperature):
        self.temperature = temperature
```

```python
house = House()
```

```python
print(house.floor)
print(house.colour_paint)
print(house.n_rooms)
```

```python
house.paint_walls('green')
house.colour_paint
```

```python
house.renovate_floor('parquet')
house.floor
```

```python
house.regulate_temp('21')
house.temperature
```
